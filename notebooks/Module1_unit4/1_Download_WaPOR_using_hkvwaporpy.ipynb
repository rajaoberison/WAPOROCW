## 1. كتالوج قاعدة البيانات/cubes
    
</div>

<div dir="rtl">
تسمح الحزمة hkvwaporpy بجلب قائمة مجموعات البيانات في قاعدة بيانات WaPOR باستخدام الوظيفة get_catalogus(). تحتوي كل مجموعة بيانات أو مكعب بيانات على رمز مكعب محدد. على سبيل المثال، رمز المكعب للمستوى 1 هطول الأمطار (يوميًا) هو L1_PCP_E. قم بتشغيل خلية التعليمات البرمجية أدناه، ثم حاول تغيير قيمة الإصدار المعلمة ل "2.0" لتحديد WaPOR الإصدار 2.0.

```python
df=hkv.read_wapor.get_catalogus(version='1.1') 
df[['caption','code']]
```

<div dir="rtl">
    
## 2.معلومات مجموعة البيانات / المكعب
    
</div>


<div dir="rtl">    
تحصل الحزمة hkvwaporpy أيضًا على معلومات مجموعة البيانات باستخدام الوظيفة
get_info_cube برمز مكعب معين. تسمى مجموعة البيانات بالمكعب لأنها تحتوي على 3 أبعاد (خط العرض وخط الطول والوقت). على سبيل المثال ، الكود المكعب للمستوى 1 Dekadal Reference EvapoTranspiration هو L1_RET_D بالنسبة لمجموعة البيانات الأخرى ، تحقق من قائمة كتالوج الرموز.    
</div>

```python
ds_code='L1_RET_D' ## كود المكعب للمستوى 1 Dekadal Reference البخرنتح
cube_info=hkv.read_wapor.get_info_cube(cube_code=ds_code)
print(cube_info)
```

<div dir="rtl">
عند طباعة معلومات cube_info ، تظهر تفاصيل مجموعة البيانات، مثل التنسيق ونوع البيانات والمدى المكاني والدقة وما إلى ذلك. لمشاهدة المزيد من التفاصيل من cube_info، استخدم cube_info.loc['detail',ds_code]. على سبيل المثال، تعطي الأبعاد معلومات حول الدقة الزمنية (الخطوة الزمنية) وتوفر المقاييس معلومات حول الوحدة والقيمة المتعددة لتحويل الرقم الرقمي (البيانات الأولية) إلى القيمة المادية.
</div>

```python
cube_info.loc['dimensions',ds_code]
```

```python
cube_info.loc['measures',ds_code]
```

<div dir="rtl">
    
## 3. توافر البيانات ومعرف البيانات النقطية

</div>

<div dir="rtl">
للحصول على قائمة بالبيانات النقطية المتوفرة في مجموعة البيانات / المكعب لفترة، استخدم الدالة get_data_availability مع معلومات cube_info التي تم الحصول عليها. سيتم تحديد كل خريطة نقطية بواسطة معرف البيانات النقطية (raster_id). على سبيل المثال، يتم استخدام الكود أدناه للحصول على المستوى 1 Dekadal الفعلي لبيانات البخرنتح والاعتراض من 2008-01-01 إلى 2018-12-31. هنا ، لاحظ أن بيانات WaPOR متاحة فقط من عام 2009.
</div>

```python
df_avail=hkv.read_wapor.get_data_availability(cube_info=cube_info,time_range='[2008-01-01,2018-12-31]')
df_avail[['raster_id']]
```

<div dir="rtl">
في قائمة data_avail المتاحة، يمكن الوصول إلى raster_id من رقم الفهرس. على سبيل المثال، النقطية الأولى مع الفهرس i = 0 لها raster_id
L1_RET_0901. يتم استخدام هذا المعرف لاحقًا لطلب عنوان URL للتنزيل من قاعدة البيانات.
</div>

```python
i=0 #احصل على أول طبقة راستر من df_avail
cube_period=df_avail.iloc[i]
raster_id=cube_period['raster_id']
print(raster_id)
```

<div dir="rtl">
    
## 4. احصل على رابط التحميل

</div>
<div dir="rtl">     
للحصول على رابط تنزيل البيانات النقطية (أو مجموعة الخرائط) ، استخدم الوظيفة get_coverage_url. تتطلب هذه الوظيفة كود adataset / cube ، ومعرف نقطي ، ورمز API صالحًا ، يتم الحصول عليه من
(https://wapor.apps.fao.org/profile)
باستخدام حساب WaPOR (راجع دفتر الملاحظات 0_Start_here (0_Start_here.ipynb) للحصول على إرشادات حول كيفية الحصول على رمز API المميز)
</div>
<div dir="rtl">
قم بتعيين قيمة متغير api_token في خلية الكود أدناه عن طريق استبدال Your_WaPOR_API_token برمز yourapi الذي تم الحصول عليه من ملف تعريف WaPOR. تذكر أن تضعه بين علامتي اقتباس مفردتين لتعريفهما كمتغير سلسلة.
</div>

```python
# احصل على رمز الAPI الخاص بك https://wapor.apps.fao.org/profile
api_token='Your_WaPOR_API_token'
```

<div dir="rtl">    
من النتائج السابقة ، ds_code هنا هو L1_RET_D و raster_id هنا هو L1_RET_0901. ستعيد الوظيفة عنوان url لتنزيل download_url مع انتهاء وقت الصلاحية expiry_datetime. إذا كان رمز API المقدم في الخلية السابقة صحيحًا ، فسيكون hkvwaporpy قادرًا على جمع الارتباط (url) لتنزيل البيانات النقطية. إذا لم يكن الأمر كذلك ، فستُرجع خلية الرمز أدناه "استجابة" KeyError ، مما يعني عدم وجود استجابة من الخادم.
</div>

```python
cov_object = hkv.read_wapor.get_coverage_url(
    APItoken = api_token, 
    raster_id = raster_id,
    cube_code = ds_code)
print(cov_object)
```

## 5. قم بتنزيل مجموعة الخرائط النقطية من المستوى 1
    
</div>
<div dir="rtl">
يمكن استخدام عنوان url الذي تم الحصول عليه من وظيفة get_coverage_url لتنزيل الخريطة النقطية مباشرةً من قاعدة البيانات باستخدام request.get. على سبيل المثال ، يتم استخدام الكود أدناه لتنزيل الخريطة النقطية L1_RET_0901.tif إلى المجلد r'.\data\L1_RET_D'.    
</div>

[**r'.\data\L1_RET_D'**](data)    

<div dir="rtl">    
    لتحديد مجلد مختلف لحفظ البيانات ، قم بتغيير قيمة المجلد output_folder.
</div>

```python
output_folder=r'.\data\L1_RET_D' #ملف لحفظ الداتا
if not os.path.exists(output_folder): #تأكد أن كان output_folder صحيحاً
    os.makedirs(output_folder) #خلق output_folder
filename=output_folder+'\{0}.tif'.format(raster_id) #حفظ ملف مع معرف نقطي كاسم ملف
#تنزيل ملف نقطي
resp=requests.get(cov_object['download_url']) 
open(filename,'wb').write(resp.content) 
print(filename)
```

## 7. تنزيل مجموعة البيانات النقطية للموقع من المستوى 2

</div>
<div dir="rtl">
لتنزيل مجموعة البيانات من المستوى 2 ، يجب تحديد رمز مجموعة البيانات / المكعب أولاً ويجب الحصول على معلومات مجموعة البيانات باستخدام الوظيفة
get_info_cube . فمثلا، L2_AETI_D هو رمز المكعب للمستوى 2 DekadalActual البخرنتح و الإعتراض، كما يمكن العثور عليه في الكتالوج من المثال 1.
</div>

```python
ds_code='L2_AETI_D'
cube_info=hkv.read_wapor.get_info_cube(cube_code=ds_code)
print(cube_info)
```

<div dir="rtl">
باستخدام الوظيفة get_data_availability كما في المثال 3 ، يمكن الحصول على قائمة الخرائط النقطية المتاحة خلال فترة زمنية. على سبيل المثال ، يتم استخدام الكود أدناه للحصول على قائمة معرّف البيانات النقطية لـ L2_AETI_D في يناير 2009. لتحديد فترة زمنية مختلفة ، قم بتغيير قيمة المعلمة time_range.
</div>

```python
df_avail=hkv.read_wapor.get_data_availability(cube_info=cube_info,time_range='[2009-01-01,2009-01-31]')
df_avail[['raster_id']]
```

<div dir="rtl">
باستخدام الوظيفة get_data_availability كما في المثال 3 ، يمكن الحصول على قائمة الخرائط النقطية المتاحة خلال فترة زمنية. على سبيل المثال ، يتم استخدام الكود أدناه للحصول على قائمة معرّف البيانات النقطية لـ L2_AETI_D في يناير 2009. لتحديد فترة زمنية مختلفة ، قم بتغيير قيمة المعلمة time_range.
</div>

```python
output_folder=r'.\data\L2_AETI_D'
if not os.path.exists(output_folder): #check إن تم عمل ملف للمخرجات
    os.makedirs(output_folder) #قم بانشاء ملف للمخرجات
for i in range(len(df_avail)): # تفقد جميع المعلومات النقطية في قائمة البيانات النقطية المتاحة
    cube_period = df_avail.iloc[i] #إحصل على الطبقة النقطية
    raster_id = cube_period['raster_id'] #هوية الطبقة النقطية
    # اجمع عنوان url لتنزيل البيانات النقطية
    cov_object = hkv.read_wapor.get_coverage_url(
        APItoken=api_token,
        raster_id = raster_id,
        cube_code = ds_code,
        loc_type = loc_type, 
        loc_code = loc_code)
    print(cov_object['download_url']) #اطبع url للتحميل
    if cov_object['expiry_datetime']>datetime.datetime.now(): #تفقد ما إذا كان عنوان url غير منتهي الصلاحية
        # إذا كان عنوان url لا يزال صالحًا ، فقم بتنزيل الملف
        resp = requests.get(cov_object['download_url'])
        open(output_folder+'\{0}.tif'.format(raster_id),'wb').write(resp.content)
        print('Saved to: '+output_folder+'\{0}.tif'.format(raster_id)) #اطبع مسار الملف الذي تم تنزيله
```
